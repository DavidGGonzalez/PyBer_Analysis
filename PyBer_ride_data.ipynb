## Plot a Pandas DataFrame and Series

```python
# Import Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Load the CSV file
pyber_ride_df = pd.read_csv("Resources/PyBer_ride_data.csv")
pyber_ride_df
```

```python
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)")
plt.show()
```

```python
# Set x-axis and tick location
x_axis = np.arange(len(pyber_ride_df))
tick_locations = [value for value in x_axis]

# Plot
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)")
plt.xticks(tick_locations, pyber_ride_df["Month"])
plt.show()
```

```python
x_axis = np.arange(len(pyber_ride_df))
x_axis
```

```python
# Bar Char
pyber_ride_df.plot.bar(x="Month", y="Avg. Fare ($USD)")
plt.show()
```

```python
# We can also simply use the plot() adding the "kind" argument with 'bar' as value
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)", kind="bar")
plt.show()
```

```python
# Skill Drill
# import dependencies

# I believe that I could use either this approach, calculate the STDEV ahead
# Or include this sentence inside the plot as an argument value
stdev = pyber_ride_df["Avg. Fare ($USD)"].std()
stdev

#pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)", color="skyblue", kind="bar", yerr=stdev)
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)", color="skyblue", kind="bar", yerr=pyber_ride_df["Avg. Fare ($USD)"].std(), capsize=3)
plt.xticks(rotation=360)
plt.yticks(np.arange(0, 51,step=5.0))

plt.show()
```
