# Line Charts

```python
# Import MatPlotLib using the pyplot module (Dependency)
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot plot(x,y); using MATLAB
plt.plot(x_axis,y_axis)
```

```python
# Create the plot plot(x,y); using Object-Oriented, we should use ax.plot() instead of plt.plot()
fig, ax = plt.subplots()
ax.plot(x_axis,y_axis)
```

```python
# try another way
fig = plt.figure()
ax = fig.add_subplot()
```

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis,y_axis)
```

```python
# Create a plot using ax.plt()
ax = plt.axes()
ax.plot(x_axis,y_axis)
```

```python
# plt.show()
plt.plot(x_axis,y_axis)
plt.show()
```

```python
# Create a plot and add a label for the legend
plt.plot(x_axis, y_axis, label='Boston')

# Create labels for the axes
plt.xlabel("Date")
plt.ylabel("Fare ($)")

# Set Y limits
plt.ylim(0, 45)

# Add a title
plt.title("PyBer Fare by Month")

# Add a legend
plt.legend()
```

```python
# Add Colors, line width, and a grid
# Create a plot and add a label for the legend
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')

# Create labels for the axes
plt.xlabel("Date")
plt.ylabel("Fare ($)")

# Set Y limits
plt.ylim(0, 45)

# Add a title
plt.title("PyBer Fare by Month")

# Add a grid
plt.grid()

# Add a legend
plt.legend()
```

```python
# Add X Limits
# Create a plot and add a label for the legend
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')

# Create labels for the axes
plt.xlabel("Date")
plt.ylabel("Fare ($)")

# Set Y limits
plt.ylim(0, 45)

# Set X Limits
# We could use
# - plt.xlim(0, 5) by list position index
# - plt.xlim("Jan", "June") by label
#plt.xlim("Jan", "June")
plt.xlim(0, 5)

# Add a title
plt.title("PyBer Fare by Month")

# Add a grid
plt.grid()

# Add a legend
plt.legend()
```

```python
# Change Figure size
# Set Figure Size, it should be set prior to ploting the points
plt.figure(figsize=(20,10))

# Create a plot and add a label for the legend
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')

# Create labels for the axes
plt.xlabel("Date")
plt.ylabel("Fare ($)")

# Set Y limits
plt.ylim(0, 45)

# Add a title
plt.title("PyBer Fare by Month")

# Add a grid
plt.grid()

# Add a legend
plt.legend()
```

```python
# Now let's do the same but using Object-Oriented approach plus a few changes
# Set object
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis, marker="x", color="green", linewidth=2, label='Boston')

# Create labels for the axes
ax.set_xlabel("Date")
ax.set_ylabel("Fare ($)")

# Add Title
ax.set_title("PyBer Fare by Month")

# Add Y Limit
ax.set_ylim(0, 45)

# Add legend
ax.legend()

# Add grids
ax.grid()
```

# Bar Charts

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create a bar chart
plt.bar(x_axis, y_axis)
```

```python
# Create a bar chart (Vertical)
plt.bar(x_axis, y_axis, color="green", label="Boston")

# Create Labels
plt.xlabel("Date")
plt.ylabel("Fare ($)")

# Create a Title
plt.title("PyBer Fare by Month")

# Add legend
plt.legend()
```

```python
# Create a bar chart (Horizontal)
# To create a horizontal bar char, simply add h at the end of bar plt.barh
plt.barh(x_axis, y_axis, color="green", label="Boston")

# Create Labels
plt.xlabel("Fare ($)")
plt.ylabel("Date")

# Create a Title
plt.title("PyBer Fare by Month")

# Add legend
plt.legend()

# Add Grids for testing only!
plt.grid()
```

```python
# Create a bar chart (Horizontal)
# To create a horizontal bar char, simply add h at the end of bar plt.barh
# Now change the data to opposite axes
plt.barh(y_axis, x_axis, color="green", label="Boston")

# Create Labels
plt.ylabel("Fare ($)")
plt.xlabel("Date")

# Create a Title
plt.title("PyBer Fare by Month")

# Add legend
plt.legend()

### NOT GOOD TO REPRESENT THIS TYPE OF DATA
```

```python
# Create a bar chart (Horizontal)
# To create a horizontal bar char, simply add h at the end of bar plt.barh
#
# Now inverting the order in which the months appear in the graph
#
plt.barh(x_axis, y_axis, color="magenta", label="Boston")

# Invert the order in which the months appear, January on top
plt.gca().invert_yaxis()

# Create Labels
plt.xlabel("Fare ($)")
plt.ylabel("Date")

# Create a Title
plt.title("PyBer Fare by Month")

# Add legend
plt.legend()
```
